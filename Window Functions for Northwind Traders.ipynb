## Question 1

Rank employees based on their total sales amount so that the management team may recognize and reward the top-performing employee as well as identifying the lowest ranked employee so that they can offer resources for improvement.

```python
%%sql
-- CTE joining three tables to get employee info and calculate their total sales.
WITH total_employee_sales AS 
(SELECT e.employee_id,
        e.first_name || ' ' || e.last_name AS full_name,
        SUM(od.quantity * od.unit_price * (1 - od.discount)) AS total_sales
   FROM employees AS e
  INNER JOIN orders AS o
     ON e.employee_id = o.employee_id 
  INNER JOIN order_details AS od
     ON o.order_id = od.order_id
  GROUP BY e.employee_id
)

SELECT *, 
       RANK() OVER(ORDER BY total_sales DESC)
  FROM total_employee_sales;
```

From the query above, it's shown that Margaret P is the top-selling employee while Steven B. is the lowest.

## Question 2

Provide visuals to the management team with a clear depiction of sales trends to identify periods of high or low sales activity.

```python
%%sql
WITH sales_progress AS
(SELECT SUM(od.quantity * od.unit_price * (1 - od.discount)) AS total_sales,
        DATE_TRUNC('month', o.order_date)::DATE AS month
   FROM orders AS o
   JOIN order_details AS od
  USING (order_id)
  GROUP BY o.order_date

)
SELECT *,
       SUM(total_sales) OVER(ORDER BY month) AS running_total
  FROM sales_progress
 ORDER BY month
 LIMIT 10;
```

## Question 3

Calculate the percentage change in sales from one month to the next using the results from the previous question.

```python
%%sql
WITH monthly_total AS
(SELECT SUM(od.unit_price * od.quantity * (1 - od.discount)) AS total_sales,
        EXTRACT(MONTH FROM o.order_date) AS month,
        EXTRACT(YEAR FROM o.order_date) AS year
   FROM order_details AS od
   JOIN orders AS o
  USING (order_id)
  GROUP BY EXTRACT(YEAR FROM o.order_date), EXTRACT(MONTH FROM o.order_date)
),
previous_total AS
(SELECT month,
        year,
        total_sales,
        LAG(total_sales) OVER(ORDER BY month) AS lag_total
   FROM monthly_total
)

SELECT year,
       month, 
       ((total_sales - lag_total) / lag_total) * 100 AS sales_growth
  FROM previous_total;
```

## Question 4

Identify customers with above_average order values.

```python
%%sql
WITH order_values AS
(
    SELECT o.customer_id, o.order_id, c.company_name,
           SUM(od.unit_price * od.quantity * (1 - od.discount)) AS total_order
      FROM customers AS c
      JOIN orders AS o
     USING (customer_id)
      JOIN order_details AS od
     USING (order_id)
     GROUP BY o.customer_id, o.order_id, c.company_name
),
orders_above_avg AS
(
SELECT order_id, customer_id , company_name, total_order,
       CASE
       WHEN total_order > AVG(total_order) OVER() THEN 'Above Average'
       ELSE 'Average'
       END as AboveAvg
  FROM order_values
)

SELECT customer_id, company_name, COUNT(*) AS no_of_orders
  FROM orders_above_avg
 WHERE AboveAvg = 'Above Average'
 GROUP BY customer_id, company_name
 ORDER BY no_of_orders DESC
 LIMIT 5;
```

We can see out top customers are Ernst Handel, Save-a-lot Markets and QUICK-Stop

## Question 5

Find the percentage of total sales for each product category

```python
%%sql
WITH product_category AS
(
    SELECT c.category_id, c.category_name,
           SUM(od.unit_price * od.quantity * (1 - od.discount)) AS total_sales
      FROM order_details AS od
      JOIN products AS p
     USING (product_id)
      JOIN categories AS c
     USING (category_id)
     GROUP BY c.category_id
)

SELECT category_id, category_name,
       total_sales / SUM(total_sales) OVER() * 100 AS pct_total_sales
  FROM product_category;
```

here we can see that pct of total sales grouped by product category with 'beverages' leading as the highest %

## Question 6

Find the top three items sold in each product category.

```python
%%sql

WITH product_category AS
(
    SELECT p.product_id, c.category_name, p.product_name,
           SUM(od.unit_price * od.quantity * (1 - od.discount)) AS total_sales
      FROM categories AS c
      JOIN products AS p
     USING (category_id)
      JOIN order_details AS od
     USING (product_id)
     GROUP BY p.product_id, c.category_name, p.product_name
),
product_ranking AS
(
    SELECT *,
       ROW_NUMBER() OVER(PARTITION BY category_name
                             ORDER BY total_sales DESC) AS product_row
  FROM product_category
 GROUP BY product_id, category_name, product_name, total_sales
)

SELECT *
  FROM product_ranking
 WHERE product_row >= 3
```
